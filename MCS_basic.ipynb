```python
import pandas as pd
import numpy as np 
import matplotlib as mpl
```

```python
#declare variables of constant values

sigma = 0.1
S0 = 100
T = 1
N = 252
K = 104
dt = T/N
M = 10000
r = 0.05
```

```python
#Simulate Price at maturity (T) using Geometric Brownian Motion

Z = np.random.standard_normal(M)
ST = S0 * np.exp((r - 0.5 * sigma**2) * T + sigma * T * np.sqrt(T) * Z)
```

```python
payoffs = np.maximum(ST - K, 0)
C0 = np.exp(-r * T) * np.mean(payoffs)
print(C0)
```
